# ## 3. Use PCA from scikit learn and elbow plot to find out reduced number of dimension (which covers more than 95% of the variance)


```python
#maping the target to integer/categorical values which can be recognised by the model
vehicle_df['class']=vehicle_df['class'].map(lambda x:0 if x=='bus' else(1 if x=='car' else 2))
```

```python
#Splitting of data
X=vehicle_df.drop(columns=['class'])
y=vehicle_df['class']
```

```python
#Standarding of data means the existing data  will be transferred to another position by using the Zscore and Zscore is varaince
from scipy.stats import zscore
X= X.apply(zscore)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=18)
pca.fit(X)
X_pca = pca.transform(X)
```

```python
display(X_pca.shape,
        np.cumsum(pca.explained_variance_ratio_))
```

```python
plt.figure(figsize=(10,7))

plt.plot(np.cumsum(pca.explained_variance_ratio_), color='k', lw=2)

plt.xlabel('Number of components')
plt.ylabel('Total explained variance')

plt.xlim(0, 20)
plt.xticks(np.arange(0, 20, 1))
plt.yticks(np.arange(0, 1.1, 0.05))

plt.axvline(21, c='b')
plt.axhline(0.95, c='r')

plt.show()
```

## From the  graph we can observe that 7 componets gives 96% of overall data variance from the dataset out of 18 features,so we can consider n_components as 7


### 4. Use Support vector machines and use grid search (try C values - 0.01, 0.05, 0.5, 1 and kernel = linear, rbf) and find out the best hyper parameters and do cross validation to find the accuracy.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3,
                                                    random_state=42)
display(X_train.shape,y_train.shape,X_test.shape,y_test.shape)
```

```python
import multiprocessing 
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
#as mentioned above considered the c values 
param_grid = [{
     'kernel': ['linear', 'rbf'],
     'C': [0.01,0.05,0.5,1.0]}]
```

```python
gs = GridSearchCV(estimator=SVC(), param_grid=param_grid,scoring='accuracy', cv=10, n_jobs=multiprocessing.cpu_count())
```

```python
# from PCA considered 7 componenst. as per the Graph
pca = PCA(n_components=7)
pca.fit(X_train)
```

```python
gs.fit(pca.transform(X_train), y_train)
```

```python
# to obtain the hyper parameters
gs.best_estimator_
```

```python
gs.best_score_ 
```

```python
best_scv=SVC(C=1.0, cache_size=200, class_weight=None, coef0=0.0,
             decision_function_shape='ovr', degree=3, gamma='auto_deprecated',
             kernel='rbf', max_iter=-1, probability=False, random_state=None,
             shrinking=True, tol=0.001, verbose=False)
best_scv.fit(pca.transform(X_train), y_train)
```

```python
prediction = best_scv.predict(pca.transform(X_test))
```

```python
x_con=pd.concat([X_test,y_test],axis=1)
x_con['predict_test']=prediction
print(X_test.shape,y_test.shape,len(prediction),x_con.shape)
x_con.head(10)
```

```python
x_con['class']=x_con['class'].map(lambda x:'bus' if x==0 else('car' if x==1 else 'van'))
x_con['predict_test']=x_con['predict_test'].map(lambda x:'bus' if x==0 else('car' if x==1 else 'van'))
pd.crosstab(x_con.predict_test,x_con['class'])
```

```python
print("Accuracy of the Test Dataset is :",best_scv.score(pca.transform(X_test), y_test))
```
