```python
import tensorflow as tf
tf.enable_eager_execution()


## REQUIRED LIBRARIES
# For data wrangling 
import numpy as np
import pandas as pd

```

```python
tf.__version__
```

```python
df = pd.read_csv('Churn_Modelling.csv', delimiter=',')
```

```python
df.shape
```

```python
df.head()
```

```python
# Check columns list and missing values
df.isnull().sum()
```

```python
# Drop the columns as mentioned which are unique
df1 = df.drop(["RowNumber", "CustomerId","Surname"], axis = 1)
```

```python
df1.head()
```

```python
#To find the Unique Values
df1.nunique()
```

```python
#Conversion of Categorical variables to numeric,as the model will consider only numeric values i.e  Geography and Gender is to be econverted
df2 = pd.get_dummies(df1, columns=["Geography","Gender"])
df2.head(10)
```

```python
#as model will accept only Float so entire dataset is to be converted into float32
df3= df2.astype('float32')
df3.head()
```

```python
#Setting X & Y
X = df3.iloc[:,:13].values
print(X)
X.shape
```

```python
y = df3.iloc[:, 13].values
print(y)
y.shape
```

```python
#Divide the data set into Train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#conversion of X train & Y train to float
X_train = X_train.astype('float32')
y_train = y_train.astype('float32')
```

```python
#Initialize Sequential model
model = tf.keras.models.Sequential()

#Normalize the data
model.add(tf.keras.layers.BatchNormalization())

#Reshaping of Data
model.add(tf.keras.layers.Dense(1000, activation='relu', input_shape=(13,)))

#Add Dense Layer which provides 10 Outputs after applying softmax
model.add(tf.keras.layers.Dense(500, activation='relu'))

model.add(tf.keras.layers.Dense(100, activation='sigmoid'))
```

```python
#Create optimizer with non-default learning rate
sgd_optimizer = tf.keras.optimizers.SGD(lr=0.05)
```

```python
#Compile the model
model.compile(optimizer=sgd_optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, 
          epochs=100,
          batch_size=10)
```

```python
#The Accuracy of the predicted values

model.evaluate(X_test,y_test)
```

```python
model.summary()
```

```python
#Note:the percentage of churn customers is about 20%.
```

```python
#Model prediction
prediction = model.predict(X_test[0:1])

#Print prediction
print(prediction)
```
